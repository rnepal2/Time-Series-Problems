### Training, Testing Datasets

```python
config = Config()
```

```python
def ECGDataset(df: pd.DataFrame=train_df, train: bool=True, augment: bool=True)-> Tuple[np.array]:
    
    y = df.values[:, -1]
    X = df.values[:, :-1]
    
    # reshaping
    X = X.reshape(X.shape[0], X.shape[1], 1)  # (# of sample, singal width(187),  1)
    y = y.reshape(-1, 1)
    
    # target label encoding
    ohe = OneHotEncoder(sparse=False)
    y = ohe.fit_transform(y)
    
    # for test data
    if not train:
        return (X, y)

    # training data
    if augment:
        Xs = []; ys = []
        
        for f, l in tqdm(zip(X, y)): 
            # if minority classes: augment each record
            if np.argmax(l) != 0:
                Xs.append(transform(f, train=True))
                ys.append(l)
            else: # else: only a few times (10%)
                if np.random.randn() < 0.1:
                    Xs.append(transform(f, train=True))
                    ys.append(l)
                
        Xs = np.array(Xs)
        ys = np.array(ys)
        
        X = np.concatenate([X, Xs])
        y = np.concatenate([y, ys])
    
    return (X, y)
```

```python
X_train, y_train = ECGDataset(train_df, train=True, augment=True)
```

```python
print("X_train.shape: ", X_train.shape)
print("y_train.shape: ", y_train.shape)
```

```python
X_test, y_test = ECGDataset(test_df, train=False)
```

```python
print("X_test.shape: ", X_test.shape)
print("y_test.shape: ", y_test.shape)
```

```python
# Number of classes and Sequence Length

print("Sequence length: ", config.seq_length)
print("Number of classes: ", config.n_class)
```
